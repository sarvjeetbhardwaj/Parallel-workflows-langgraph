```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
from langchain_google_genai  import ChatGoogleGenerativeAI

load_dotenv()


llm =  ChatGoogleGenerativeAI(
    model="gemini-3.1-flash-lite",
    temperature=0)
```

```python
class EmployeeStete(TypedDict):
    employee_name : str
    monthly_salary: int
    working_days: int
    completed_projects: int

    yearly_salary: int
    bonus_amount: int
    project_status: str
    summary: str
    
```

```python
def calculate_yearly_salary(state:EmployeeStete)->EmployeeStete:
    yearly_salary = state['monthly_salary'] * 12
    state['yearly_salary'] = yearly_salary
    return state

def calculate_bonus(state:EmployeeStete)->EmployeeStete:
    bonus = state['yearly_salary'] * 0.2
    state['bonus_amount'] = bonus
    return state

def project_evaluation(state:EmployeeStete)->EmployeeStete:
    if state['completed_projects'] >=5:
        status = 'Excellent'
    else:
        status = 'Average'

    state['project_status'] = status
    return state
def summary(state:EmployeeStete)->EmployeeStete:
    summary_text = f'Employee {state["employee_name"]} has an yearly salary of {state['yearly_salary']} with bonus \
                    {state['bonus_amount']} with status as {state['project_status']}'
    
    state['summary'] = summary_text

    return state


```

```python
graph = StateGraph(EmployeeStete)

graph.add_node('calculate_yearly_salary', calculate_yearly_salary)
graph.add_node('calculate_bonus', calculate_bonus)
graph.add_node('project_evaluation', project_evaluation)
graph.add_node('summary', summary)


```
